```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
population_file = "Resources/US_Population_Data.csv"
population_df = pd.read_csv(population_file)
population_df.head()
```

```python
population_df["STATE"] = population_df["STATE"].astype("string")
population_df.dtypes
```

```python
population_df.dtypes
```

```python
minimum_wage_file = "Resources/Minimum_Wage_Data.csv"
minimum_wage_df = pd.read_csv(minimum_wage_file)
minimum_wage_df.head()
```

```python
minimum_wage_df.head()
```

```python
minimum_wage_df["State"] = minimum_wage_df["State"].astype("string")
minimum_wage_df.dtypes
```

```python
minimum_wage_df.State = minimum_wage_df.State.apply(lambda x: str(x))
minimum_wage_df.dtypes
```

```python
#Created a filtered Dataframe from specific columns

population_cols = ["STATE", "POPESTIMATE2019"]
population_transformed = population_df[population_cols]

population_transformed = population_transformed.rename(columns={"STATE": "state", 
                                                                "POPESTIMATE2019": "population_estimate"})
```

```python
#population_transformed.set_index("State", inplace=True)

population_transformed.head()
```

```python
minimum_wage_cols = ["Year", "State", "State.Minimum.Wage"]
minimum_wage_filtered = minimum_wage_df[minimum_wage_cols]
```

```python
minimum_wage_filtered.head()
```

```python
minimum_wage_transformed = minimum_wage_filtered.loc[minimum_wage_filtered["Year"] == 2019, : ]
```

```python
minimum_wage_transformed = minimum_wage_transformed.rename(columns={"State.Minimum.Wage": "state_minimum_wage",
                                                                   "State": "state",
                                                                   "Year":"_year"})
minimum_wage_transformed
```

```python
final_df = pd.merge(minimum_wage_transformed, population_transformed, on="state")
```

```python
final_df.drop_duplicates("state", inplace=True)
```

```python
final_df = final_df.rename(columns={"year": "_year",
                                   "State_Minimum_wage": "state_minimum_wage"
                                   })
final_df
```

```python
final_df.dtypes
```

```python
connection_string = "postgres:postgres@localhost:5433/state_db"
engine = create_engine(f'postgresql://{connection_string}')
engine
```

```python
engine.table_names()
```

```python

minimum_wage_transformed.to_sql(name='wage', con=engine, if_exists='append', index=False)
population_transformed.to_sql(name='population', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from wage', con=engine).head()
```

```python
pd.read_sql_query('select * from population', con=engine).head()
```
